## 必要なモジュールをインポートする

```python
import numpy as np
import matplotlib.pyplot as plt
```

## データを定義する

```python
NUMBERS = 100
rng = np.random.default_rng()

x, y = rng.random(NUMBERS), rng.random(NUMBERS) # For scatter plots
x_std = rng.standard_normal(NUMBERS) # For histogram
data = rng.integers(low=50, high=101, size=20) # For box plot
```

## グラフを描画する

### Object-oriented style の場合

```python
plt.style.use('seaborn')
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

ax1.scatter(x, y)
ax2.hist(x_std)
ax3.boxplot(data)

plt.show()
```

### Pyplot-style の場合

```python
plt.style.use('seaborn')
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(x, y)

plt.subplot(1, 3, 2)
plt.hist(x_std)

plt.subplot(1, 3, 3)
plt.boxplot(data)

plt.show()
```
